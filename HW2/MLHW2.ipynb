```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import linalg


#定義φ,w,t,φ_transpose,第六題的時候把維度改成只有前面兩個維度就好了
df = pd.read_csv('train.csv',header=None)
df2 = pd.read_csv('test.csv',header=None)
φ_test = df2.values
t = df.iloc[0:180,0:3]
t = t.values
t = np.transpose(t)
φ = df.iloc[0:180,3:11]
w = df.iloc[0:3,3:11]
φ = φ.values
φ_transpose = np.transpose(φ)

#將W設成0,第六題的時候把維度改成2
for i in range(0,2):
    for j in range(0,7):
        w.iloc[i:j] = 0      
w = w.values

  
#定義error function  
def error(φ,t,w):
    sum = 0
    φ_transpose = np.transpose(φ)
    a = np.dot(w,φ_transpose)    
    y = np.dot(w,φ_transpose)
    for j in range(0,3):
        for n in range(0,180):
            y[j,n] = math.exp(a[j,n])/(math.exp(a[0,n])+math.exp(a[1,n])+math.exp(a[2,n]))
    for j in range(0,3):
        for n in range(0,180):
            sum = sum+t[j,n]*math.log(y[j,n])
    sum = -sum
    return sum



EW = []
accuracy = []

#迭代求解W
while error(φ,t,w)>0.01:
    φ_transpose = np.transpose(φ)
    a = np.dot(w,φ_transpose)    
    y = np.dot(w,φ_transpose)
    summ = 0
    #迭代
    for j in range(0,3):
        for n in range(0,180):
            y[j,n] = math.exp(a[j,n])/(math.exp(a[0,n])+math.exp(a[1,n])+math.exp(a[2,n]))
    HJ = [0,0,0]
    for i in range(0,3):
        #第六題的時候把DE改成2維度
        DE = [0,0,0,0,0,0,0]
        
        for n in range(0,180):  
            DE = DE+(y[i,n]-t[i,n])*φ[n]
        for n in range(0,180):
            φn_df = pd.DataFrame(φ[n])
            φn_df_trans = φn_df.transpose()
            s = np.dot(φn_df,φn_df_trans)
            HJ[i] = HJ[i] +(s*y[i,n]*(1-y[i,n]))

        HJ[i] = np.transpose(HJ[i])
        HJinv = np.linalg.inv(HJ[i])
        w[i] = w[i]-np.dot(DE,HJinv)
    EW.append(error(φ,t,w))
    
    #計算accuracy
    for n in range(0,180):
        max = 0
        sit = 0
        for j in range(0,3):
            if y[j,n]>max:
                max = y[j,n]
                sit = j
        if t[sit,n] == 1:
             summ = summ+1
    summ = summ/180
    accuracy.append(summ)
    
#求test的classification result
φ_test_transpose = np.transpose(φ_test)
a = np.dot(w,φ_test_transpose)    
y = np.dot(w,φ_test_transpose)
for j in range(0,3):
    for n in range(0,30):
        y[j,n] = math.exp(a[j,n])/(math.exp(a[0,n])+math.exp(a[1,n])+math.exp(a[2,n]))
for n in range(0,30):
    max = 0
    sit = 0
    for j in range(0,3):
        if y[j,n]>max:
            max = y[j,n]
            sit = j
    for j in range(0,3):
        if sit == j:
            y[j,n] = 1
        else:
            y[j,n] = 0

#plot the histogram&2D graph
histogram = []
histogram2 = []
histogram3 = []
Dgraphx = []
Dgraphy = []
Dgraphx2 = []
Dgraphy2 = []
Dgraphx3 = []
Dgraphy3 = []
#代入0~6dimension求各dimension的histogram
for n in range(0,60):
    histogram.append(φ[n,0])
for n in range(60,120):
    histogram2.append(φ[n,0]) 
for n in range(120,180):
    histogram3.append(φ[n,0])
#選第一個跟第二個dimension畫 2D graph
for n in range(0,60):
    Dgraphx.append(φ[n,0])
    Dgraphy.append(φ[n,1])
for n in range(60,120):
    Dgraphx2.append(φ[n,0])
    Dgraphy2.append(φ[n,1])
for n in range(120,180):
    Dgraphx3.append(φ[n,0])
    Dgraphy3.append(φ[n,1])
    
#畫histogram    
#plt.hist([histogram, histogram2 ,histogram3], stacked=True)
#畫EW 
#plt.plot(EW)
#畫2D graph
#plt.scatter(Dgraphx,Dgraphy)
#plt.scatter(Dgraphx2,Dgraphy2)
#plt.scatter(Dgraphx3,Dgraphy3)

```
